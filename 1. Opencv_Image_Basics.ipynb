# Flipping The Image

1. Horizontal Flipping  :-- Value 0
2. Vertical Flipping :- Value 1
3. Both Directional Flipping :- Value -1

```
figure = plt.figure(figsize=(20,15))
ax = figure.add_subplot(221)
ax.set_title("Original Image")
ax.imshow(rgbImage)
ax.grid(False)
ax = figure.add_subplot(222)
ax.set_title("Horizontal Flipped Image")
ax.imshow(cv2.flip(rgbImage,0))
ax.grid(False)
ax = figure.add_subplot(223)
ax.set_title("Vertical Flipped Image")
ax.imshow(cv2.flip(rgbImage,1))
ax.grid(False)
ax = figure.add_subplot(224)
ax.set_title("Both Directional Flipped Image")
ax.imshow(cv2.flip(rgbImage,-1))
ax.grid(False)
```

# Drawing Rectangle, Circle And Line,Polygons and Writing Text On Image

```
import numpy as np
```

```
blackImage = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```
plt.imshow(blackImage)
plt.grid(False)
```

```
cv2.rectangle(img=blackImage,pt1=(10,10),pt2=(100,100),color=(0,255,0),thickness=10)
cv2.circle(img=blackImage,center=(300,300),radius=50,color=(255,0,0),thickness=-1)
cv2.line(img=blackImage,pt1=(0,0),pt2=(500,500),color=(0,0,255),thickness=10)
cv2.putText(img=blackImage,org=(50,400),text="Panther",fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(255,255,255),thickness=5)
vertices = np.array([[100,200],[200,300],[300,300]],dtype=np.int32)
pts = vertices.reshape((-1,1,2))
cv2.polylines(blackImage,[pts],color=(255,255,255),thickness=10,isClosed=True)

showImage(blackImage)
```
